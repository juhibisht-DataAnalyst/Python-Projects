**Exploratory Data Analysis**

**Gender**

```python
df.columns
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax = sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Gender',y='Amount',data= sales_gen)
```

**From above graph we can see that most of the buyers are female and even the purchasing power of female are greater than male**

**Age**

```python
ax= sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,color='Green')
```

**From above graph we can see that most of the buyers are age group between 26-35 yrs female**

**State**

```python
#Total numbers of orders from top 10 states

sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state,color='Pink')
```

```python
#Total amount/sales from top 10 states
sales_amount= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_amount,x='State',y='Amount',color='Pink')
           
```

**From above graph we can see that most of the orders & total amount are from Uttar Pradesh,Maharastra and Karnataka respectively.**

**Marital Status**

```python
ax = sns.countplot(data=df,x='Marital_Status',color='Orange')
for bars in ax.containers:
    ax.bar_label(bars)

sns.barplot(rc={'figure.figsize':(5,6)})
```

```python
sales_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_status,x='Marital_Status',y='Amount',hue='Gender')
```

**From above graph we can see that most of the buyers are married(women) and they have high purchasing power.**

**Occupation**

```python
ax = sns.countplot(data=df,x='Occupation',color='Red')

for bars in ax.containers:
    ax.bar_label(bars)

sns.set(rc={'figure.figsize':(15,10)})
```

```python
sales_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount')

sns.barplot(x='Occupation',y='Amount',data=sales_occupation,color='Red')
sns.set(rc={'figure.figsize':(6,5)})
```

**From above graph we can see that most of the buyers are working in IT Sector, Healthcare, Aviation.**

```python
df.columns
```

**Product Category**

```python
ax = sns.countplot(data=df,x='Product_Category',color='Yellow')

for bars in ax.containers:
    ax.bar_label(bars)

sns.set(rc={'figure.figsize':(30,15)})
```

```python
sales_product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()

sns.barplot(x='Product_Category',y='Amount',data=sales_product,color='Yellow')
sns.set(rc={'figure.figsize':(15,5)})
```

*From above graphs we can see that most of the sold products are Food, Clothing & Apparel, Electronics and Footwear & shooes.* 

```python
#Top 10 Product Orders
sales = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(x='Product_ID',y='Orders',data=sales,color='Brown')
sns.set(rc={'figure.figsize':(20,5)})
```

**Conclusion:**

**Maried women age group 26-35 yrs from UP, Maharastra and Karnataka working IT,Healthcare and Aviation are more highly to buy products from Food, Clothing & Electronics**
