**Data Visualization**

```python
#Size barplot
ax = sns.countplot(x='Size',data=df)
#Check the data labels
for bars in ax.containers:
    ax.bar_label(bars)
```

**Most of the people buys M-Size.**

```python
#Courier Status	
plt.figure(figsize=(10,6))
sns.countplot(x='Courier Status',hue='Status',data=df)
plt.show()
```

**Majority to the orders are Shipped through the Courier Status.**

```python
df['Category'] = df['Category'].astype('str')
```

```python
df['Category'].dtype
```

```python
category_data = df['Category']
plt.figure(figsize=(10,6))
plt.hist(category_data,bins=12,edgecolor='Black',color='Yellow')
plt.xticks(rotation=90)
plt.show()
```

**Most of the buyers are T-Shirts.**

```python
#Check b2b data
check_B2B=df['B2B'].value_counts()

#Create a pie chart
plt.pie(check_B2B, labels=check_B2B.index, autopct='%1.1f%%')
plt.show()
```

**Maximum 99.2% of buyers are retailers and 0.8% are wholesales.**

```python
#Scatter plot
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available Size')
plt.show()
```

```python
#Distribution wise ship-state and Top 10 States
top10_state = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(15,6))
sns.countplot(data=df[df['ship-state'].isin(top10_state.index)],x='ship-state')
plt.xlabel('State')
plt.ylabel('Order_count')
plt.title('Distribution of States')
plt.show()
```

**Most of the buyers are Maharashtra state.**
